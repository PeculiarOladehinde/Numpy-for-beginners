**MATHS OPERATIONS**

```python
# adding two numpy array values
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = a+b
result
```

```python
# This is another way of adding numpy arrays together using the add() method

np.add(a,b)
```

```python
# subtracting a numpy array value from another numpy array value
a = np.array([10, 20, 30])
b = np.array([5, 5, 15])

result = a-b
result
```

```python
np.subtract(a,b)
```

```python
num_ = np.array([1, 2, 3])
num_1 = np.array([4, 5, 6])

solu = num_ * num_1
solu
```

```python
np.multiply(num_, num_1)
```

```python
a = np.array([10, 20, 30])
b = np.array([5, 5, 15])

result = a / b
result
```

```python
a = np.array([10, 20, 30])
b = np.array([5, 5, 15])

result = a / b
print(result.astype(int))
```

**Exponentiation(Power) in NumPy**
